```python
#data cleaning
import pandas as pd
import numpy as np
df = pd.read_csv("/content/drive/MyDrive/Hotel Bookings (1) (1).csv")
df.head()


df['children'] = df['children'].fillna(0)
df['country'] = df['country'].fillna("Unknown")
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)
```

```python
#remove duplicates
df.duplicated().sum()

df = df.drop_duplicates()
```

```python
 #createing new column
df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
```

```python
#hotel type analysis
sns.countplot(x='hotel', data=df)
plt.title("Hotel Type Distribution")
plt.show()
```

```python
#Booking per Year
sns.countplot(x='arrival_date_year', data=df)
plt.title("Bookings per Year")
plt.show()
```

```python
#Booking by month
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/drive/MyDrive/Hotel Bookings (1) (1).csv")

plt.figure(figsize=(12,4))
sns.countplot(x='arrival_date_month', data=df)

plt.xticks(rotation=30)
plt.title("Bookings by Month")

plt.show()
```

```python
#Adr Analysis
plt.figure(figsize=(7,2))
sns.boxplot(x='hotel', y='adr', data=df)

plt.title("ADR by Hotel Type", fontsize=15)
plt.xlabel("Hotel Type", fontsize=14)
plt.ylabel("Average Daily Rate (ADR)", fontsize=10)

plt.show()
```

```python
#costmer type analysis
sns.countplot(x='customer_type', data=df)
plt.xticks(rotation=45)
plt.title("Customer Types")
plt.show()
```

```python
#market analysis
plt.figure(figsize=(8,4))
sns.countplot(x='market_segment', data=df)
plt.xticks(rotation=45)
plt.title("Market wise Distribution")
plt.show()
```

```python
#Top country by booking
top_country = df['country'].value_counts().head(10)

top_country.plot(kind='bar')
plt.title("Top 10 Countries Booking Hotels")
plt.show()
```

```python
#correlation heatmap
plt.figure(figsize=(10,2))
sns.heatmap(df.corr(numeric_only=True), annot=False)
plt.title("Correlation Heatmap")
plt.show()
```
